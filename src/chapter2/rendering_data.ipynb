```python
import pandas as pd
from PIL.ImageColor import colormap
from matplotlib.pyplot import xticks
from numpy.array_api import trunc
from scipy.stats import alpha

housing = pd.read_csv("../../datasets/housing/housing.csv")
housing.head(100)
```

```python
housing.plot(x="longitude", y="latitude", kind="scatter", grid=True, alpha=0.3)
```

```python
housing.plot(x="longitude", y="latitude", kind="scatter", grid=True, alpha=.5,
             s=housing["population"]/100, 
             c="median_house_value", cmap="magma", colorbar=True,
             )

```

```python
# housing["median_house_scaled"] = housing["median_house_value"].apply(lambda x: x/10)

housing.plot(kind="scatter", x="median_income", y="median_house_value",
             alpha=.1, grid=True, 
             )
```

```python
from pandas.plotting import scatter_matrix

housing["bedroom_per_house"] = housing[["total_bedrooms","households"]].apply(lambda x: x["total_bedrooms"] / x["households"], axis=1)
housing["room_per_house"] = housing[["total_rooms","households"]].apply(lambda x: x["total_rooms"] / x["households"], axis=1)


housing.plot(kind="scatter", x="room_per_house", y="median_house_value",
             alpha=.1, grid=True, 
            )
housing["population_scaled_down"] = housing["population"].apply(lambda x: x/100)
housing.plot(kind="scatter", x="population_scaled_down", y="median_house_value",
             alpha=.1, grid=True, 
            )
```
